```python
# your code here
def train_model(model,train_loader,val_loader,loss_fn,optimizer,epochs: int, device=device):
    losses_train = []
    losses_test = []
    acc = []
    for epoch in range(epochs):
        start_time = time.time()
        model.train(True)
        for X,Y in train_loader:
            X,Y = X.to(device), Y.to(device)
            output=model(X)
            output=output.T[0]
            loss=loss_fn(output, Y)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            losses_train.append(loss.item())
        model.train(False)
        with torch.no_grad():
            for X,Y in val_loader:
                X,Y = X.to(device), Y.to(device)
                output=model(X)
                loss=loss_fn(output[0], Y)
                losses_test.append(loss.item())
                y_pred = output.max(-1)[1]
                acc.append((y_pred==Y).to(default_dtype).mean().item())
        print(f'epoch {epoch+1}/{epochs}, {time.time()-start_time:.3f}s.')
        print(f'\t training loss: {losses_train[-1]:.6f}')
        print(f'\t validation loss: {losses_test[-1]:.6f}')
        print(f'\t val accuracy: {acc[-1]:.6f}')
optimizer = torch.optim.Adam(model_task_1.parameters(), lr=1e-3)
epochs=300
train_model(model_task_1, train_loader,validation_tensor,nn.BCELoss(),optimizer,epochs,device)
```

```python
import math
from collections import Counter
def conditional_entropy(x, y, log_base: float = 2):
  y_counts = Counter(y)
  xy_counts = Counter(list(zip(x, y)))
  total_counts = len(x)
  cond_entropy = 0.0
  for xy in xy_counts.keys():
      p_xy = xy_counts[xy] / total_counts
      p_y = y_counts[xy[1]] / total_counts
      cond_entropy += p_xy * math.log(p_y / p_xy, log_base)
 
  return cond_entropy
# объявим функцию
def entropy(x, log_base: float = 2):
  x_counts = Counter(x)
  total_counts = len(x)
  p_x = list(map(lambda n: n / total_counts, x_counts.values()))
  entropy = 0.0
  for p in p_x:
    entropy += -p * math.log(p, 2)
 
  return entropy

# сразу объявим функцию
def ucoef(x, y, log_base = 2):
 
  # найдем условную энтропию S(X,Y)
  s_xy = conditional_entropy(x, y, log_base)
 
  # энтропию S(X)
  s_x = entropy(x, log_base)
 
  # подставим эти значения в формулу
  u = (s_x - s_xy) / s_x
 
  # выведем результат
  return u
for label in data.columns.tolist():
  print(f"{label}/target Ucoef: {ucoef(data['target'], data[label])}")
```

```python
import numpy as np 
cov_mat = np.cov(data.T) 
eigen_vals, eigen_vecs = np.linalg.eig(cov_mat) 
print('\nСобственные значения \n%s' % eigen_vals)
```

```python
sns.heatmap(cov_mat)
```

```python
tot = sum(eigen_vals)
var_exp =  [ (i / tot) for i in sorted(eigen_vals, reverse=True)] 
cum_var_exp = np.cumsum(var_exp) 
import matplotlib.pyplot as plt 
plt.bar(range(1,17), var_exp, alpha=0.5, align='center', label='индивидуальная объясненная дисперсия') 
plt.step(range(1,17), cum_var_exp, where='mid', label='кумулятивная объясненная дисперсия') 
plt.ylabel('Коэффициент объясненной дисперсии') 
plt.xlabel ('Индекс главного компонента') 
plt.legend(loc='best') 
plt.tight_layout() 
plt.show()
```

```python
 #создать список кортежей(собственное значение, собственный вектор) 
eigen_pairs = [(np.abs(eigen_vals[i]), eigen_vecs[:, i]) for i in range(len (eigen_vals))] 
 # отсортировать кортежи (собственное значение, >>> # собственный вектор) от высоких к низким 
eigen_pairs.sort(key=lambda k: k[0], reverse=True) 
```
