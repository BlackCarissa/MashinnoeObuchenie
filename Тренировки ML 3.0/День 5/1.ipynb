### Создаем модельку и обучаем ее. 

```python
# use GPU if available
device = torch.device("cpu")
device
```

```python
m = nn.LogSigmoid()
input = torch.randn(2)
output = m(input)
output
```

```python
# Creating model instance
default_dtype = torch.float32
model_task_1 = nn.Sequential(
    nn.Linear(16,1),
    nn.Sigmoid()

)
model_task_1.to(device,default_dtype)
```

```python
# do not change the code in the block below
# __________start of block__________
random_batch = next(iter(train_loader))
assert model_task_1 is not None, "Please, use `model_task_1` variable to store your model"

try:
    x = random_batch[0].to(device)
    y = random_batch[1].to(device)

    # compute outputs given inputs, both are variables
    y_predicted = model_task_1(x)
except Exception as e:
    print("Something is wrong with the model")
    raise e


assert y_predicted.shape[-1] == 1, "Model should predict 1 logits/probas"

print("Everything seems fine!")
# __________end of block__________
```

```python
# your code here
def train_model(model,train_loader,val_loader,loss_fn,optimizer,epochs: int, device=device):
    losses_train = []
    losses_test = []
    acc = []
    for epoch in range(epochs):
        start_time = time.time()
        model.train(True)
        for X,Y in train_loader:
            X,Y = X.to(device), Y.to(device)
            output=model(X)
            output=output.T[0]
            loss=loss_fn(output, Y)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            losses_train.append(loss.item())
        model.train(False)
        with torch.no_grad():
            for X,Y in val_loader:
                X,Y = X.to(device), Y.to(device)
                output=model(X)
                loss=loss_fn(output[0], Y)
                losses_test.append(loss.item())
                y_pred = output.max(-1)[1]
                acc.append((y_pred==Y).to(default_dtype).mean().item())
        print(f'epoch {epoch+1}/{epochs}, {time.time()-start_time:.3f}s.')
        print(f'\t training loss: {losses_train[-1]:.6f}')
        print(f'\t validation loss: {losses_test[-1]:.6f}')
        print(f'\t val accuracy: {acc[-1]:.6f}')
optimizer = torch.optim.Adam(model_task_1.parameters(), lr=1e-3)
epochs=1000
train_model(model_task_1, train_loader,validation_tensor,nn.BCELoss(),optimizer,epochs,device)
```
