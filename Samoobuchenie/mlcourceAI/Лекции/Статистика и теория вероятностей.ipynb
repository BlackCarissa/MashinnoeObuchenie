## Задача 1. Постройте диаграмму рассеяния на основе данных

Оля хочет выяснить, зависит ли средний балл за экзамен от времени его проведения. Она собрала данные за прошлый год и свела их в таблицу.

```python
urok = ['Английский','Математика','История','Информатика','ИЗО','Биология','Физика','Химия','Испанский']
vremya = [1,6,2,4,4,1,3,1,3]
ball = [93,87,70,62,86,73,73,80,96]
```

```python
import seaborn as sns
import pandas as pd

sns.set()

# Graphics in SVG format are more sharp and legible
%config InlineBackend.figure_format = 'svg'
```

```python
data = pd.DataFrame()
data['Время'] = vremya
data['Баллы'] = ball
```

```python
sns.jointplot(x='Время',y='Баллы',data=data,kind='scatter')
```
